Calculate the lengths of each record in 'review' column and add to the dataframe

```python
df['Review_Length'] = df['Review'].apply(len)
df
```

Calculate the number of words in 'review' column and add to the dataframe

```python
df['Word_Count'] = df['Review'].apply(lambda x: len(x.split()))
```

```python
df
```

Calculate the average length of word for each record in 'review' column and add to the dataframe

```python
df['Avg_Word_Length'] = df.apply(lambda row: row['Review_Length'] / row['Word_Count'], axis=1)
df
```

Calculate the average sentence length of the sentences in the 'review' and add to the dataframe


```python
df['Sentence_Count'] = df['Review'].apply(lambda x: x.count("."))
df['Avg_Sentence_Length'] = df['Word_Count'] / (df['Sentence_Count'] + 1)
df
```

Visualise each of the columns with respect to the 'Rating' column

```python
import matplotlib.pyplot as plt
plt.scatter(df['Rating'], df['Review_Length'])
plt.title('Review Length vs Rating')
plt.xlabel('Rating')
plt.ylabel('Review Length')
plt.show()

plt.scatter(df['Rating'], df['Word_Count'],color='y')
plt.title('Word_Count vs Rating')
plt.xlabel('Rating')
plt.ylabel('Word_Count')
plt.show()

plt.scatter(df['Rating'], df['Avg_Word_Length'],color='g')
plt.title('Avg_Word_Length vs Rating')
plt.xlabel('Rating')
plt.ylabel('Avg_Word_Length')
plt.show()

plt.scatter(df['Rating'], df['Sentence_Count'],color='b')
plt.title('Sentence_Count vs Rating')
plt.xlabel('Rating')
plt.ylabel('Sentence_Count')
plt.show()

plt.scatter(df['Rating'], df['Avg_Sentence_Length'],color='r')
plt.title('Avg_Sentence_Length vs Rating')
plt.xlabel('Rating')
plt.ylabel('Avg_Sentence_Length')
plt.show()
```

 Preprocess review  to lowercase

```python
df['Review'] = df['Review'].apply(lambda x: x.lower())
df
```

 Remove punctuations

```python
import string
from nltk.corpus import stopwords

df['Review'] = df['Review'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df
```

Stopwords

```python
import nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
df['Review'] = df['Review'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df
```

Store the text in the 'review' column in a form list

```python
df['Review_Lists'] = df['Review'].apply(lambda x: x.split())
df
```

Identify most common words using 'Review_lists' column


```python
import nltk
from nltk import FreqDist
all_words = [word for sublist in df['Review_Lists'] for word in sublist]
freq_dist=FreqDist(all_words)
freq_dist.most_common(10)
```

Plot the most commonly occuring words


```python
plt.figure(figsize=(10, 6))
freq_dist.plot(5, cumulative=False)

plt.show()
```
