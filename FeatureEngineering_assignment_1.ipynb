replace the NA be mean/median/mode of that column


```python
df.fillna(df.mode().iloc[0])
```

 replace the NA by values that are at the far end of the distribution of that column


```python
import numpy as np
for column in df.select_dtypes(include=[np.number]).columns:
  far_end_value=df[column].mean()+3*df[column].std()
  df[column].fillna(far_end_value, inplace=True)

# df.head(378)
```

Replace the NA with random sampling from the pool of available observation of the column


```python
df3=pd.read_csv("./vgsales.csv")
def replace_with_random_sample(x):
    if pd.isna(x):
        return np.random.choice(df3[column][df3[column].notna()])
    else:
        return x

for column in df3.columns:
    df3[column] = df3[column].apply(replace_with_random_sample)

df3.head(179)
```

Identify outliers by IQR - Interquartile Range Rule

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outliers_iqr = df[((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
outliers_iqr
```

Identify outliers by Mean and Standard Deviation

```python
mean = df.mean()
std_dev = df.std()

upper_bound = mean + 3 * std_dev
lower_bound = mean - 3 * std_dev

outliers = df[((df< lower_bound) | (df > upper_bound)).any(axis=1)]
outliers
```

Identify outliers by Median and Median Absolute Deviation


```python
median = df.median()
mad = (df - median).abs().median()

upper_bound_mad = median + 3 * mad
lower_bound_mad = median - 3 * mad

outliers = df[((df < lower_bound_mad) | (df > upper_bound_mad)).any(axis=1)]
outliers
```

Impute outliers by arbitrary value

```python
arbitrary_value = 999999

def replace_outliers_with_value(df, arbitrary_value):
    for column in df.select_dtypes(include=[np.number]):
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df[column] = df[column].mask((df[column] < lower_bound) | (df[column] > upper_bound), arbitrary_value)

    return df

df_modified = replace_outliers_with_value(df.copy(), arbitrary_value)
df_modified
```

Winsorization

```python
from scipy.stats.mstats import winsorize

data_winsorized = df.copy()

for col in df.columns:
     if pd.api.types.is_numeric_dtype(df[col]):
        data_winsorized[col] = winsorize(df[col], limits=[0.05, 0.05])
     else:
        data_winsorized[col] = df[col]


data_winsorized
```

Group the observations that show rare labels into a unique category


```python
threshold = 0.05
data_grouped = df.copy()
for column in data_grouped.columns:
    counts = data_grouped[column].value_counts()
    rare_labels = counts[counts < threshold].index
    data_grouped[column].replace(rare_labels, data_grouped[column].mode()[0], inplace=True)

data_grouped
```

replace the rare label by most frequent label


```python
data_grouped = df.copy()
for column in data_grouped.columns:
    counts = data_grouped[column].value_counts()
    rare_labels = counts[counts < threshold].index
    data_grouped[column].replace(rare_labels, data_grouped[column].mode()[0], inplace=True)

data_grouped
```
