```python
import pandas as pd
import numpy as np


df=pd.read_csv("./tested.csv")
df.dropna(inplace=True)
df.head()
```

Logarithmic Transformation

```python
df['Fare_log'] = np.log(df['Fare'])

print(df[['Fare', 'Fare_log']].head())
```

Reciprocal Transformation

```python
df['Fare_reciprocal'] = 1 / df['Fare']
print(df[['Fare', 'Fare_reciprocal']].head())
```

Square root transformation

```python
df['Fare_sqrt'] = np.sqrt(df['Fare'])
print(df[['Fare', 'Fare_sqrt']].head())
```

Exponential Transformation

```python
df['Fare_exp'] = np.exp(df['Fare'])

print(df[['Fare', 'Fare_exp']].head())
```

Box-Cox Transformation

```python
from scipy.stats import boxcox
fare_column = df["Fare"]
fare_column_positive = fare_column + abs(fare_column.min()) + 1e-6
fare_boxcox, _ = boxcox(fare_column_positive)
result_boxcox = pd.DataFrame({'Fare_BoxCox': fare_boxcox})
result_boxcox
```

Quantile Transformation

```python
from sklearn.preprocessing import quantile_transform

fare_data = df['Fare'].values.reshape(-1, 1)
df['Fare_quantile'] = quantile_transform(fare_data, n_quantiles=len(df), output_distribution='uniform', copy=True)
print(df[['Fare', 'Fare_quantile']].head())
```
