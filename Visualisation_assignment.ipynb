Loading the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('./tested.csv')
```

Show dataset

```python
df
```

Get the dtypes for each columns and  general data description of the entire dataset

```python
print(df.dtypes)
print(df.describe())
```

 barplot of a discrete variable x=pclass against target variable y=survived

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Barplot of pclass against survived')
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.show()
```

countplot of a discrete variable x=pclass

```python
sns.countplot(x='Pclass', data=df)
plt.title('Countplot of pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

boxplot of a discrete variable x=pclass against y=fare

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Boxplot of pclass against fare')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

distplot of a continuous variable x=fare


```python
sns.distplot(df['Fare'])
plt.title('Distplot of fare')
plt.xlabel('Fare')
plt.show()
```

scatter plot of two variables, x=fare and y=pclass


```python
sns.scatterplot(x='Fare', y='Pclass', data=df)
plt.title('Scatter plot of fare against pclass')
plt.xlabel('Fare')
plt.ylabel('Pclass')
plt.show()
```

correlation plot between variable

```python
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation Plot')
plt.show()
```

heatmap

```python
sns.heatmap(df.corr(), annot=True)
plt.title('Heatmap')
plt.show()
```
