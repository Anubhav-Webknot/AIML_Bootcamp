```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv("./movies.csv")
ratings_df = pd.read_csv("./ratings.csv")
movies_df.dropna( inplace=True)
ratings_df.dropna( inplace=True)
```

```python
movies_df.head()
```

```python
ratings_df.head()
```

Seprating the year column

```python
movies_df['year'] = movies_df.title.str.extract('(\d\d\d\d)',expand=False)
movies_df.head(3)
```

Removing parentheses

```python
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
movies_df.head(3)
```

Removing the years from title

```python
movies_df['title'] = movies_df.title.str.replace('(\d\d\d\d)', '')
movies_df.head(3)
```

```python
movies_df['title'] = movies_df['title'].str.replace(r"\(.*\)", "").str.strip()
movies_df
```

Applying the strip function to remove ending whitespaces

```python

movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

Every genre  is calling the split function on |


```python
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
movies_df.isna().sum()
```

fill movies_df missing year values with 0.

```python
movies_df.year.fillna(0, inplace=True)
```

one hot encoding the genre column

```python
movies_with_genres = movies_df.copy(deep=True)
x = []
for index, row in movies_df.iterrows():
    x.append(index)
    for genre in row['genres']:
        movies_with_genres.at[index, genre] = 1
print(len(x) == len(movies_df))

movies_with_genres = movies_with_genres.fillna(0)
movies_with_genres.head(3)
```

Dropping the timestamp column


```python
ratings_df.drop('timestamp', axis=1, inplace=True)
ratings_df.head(3)
```

Check for missing values

```python
ratings_df.isna().sum()
```

#Content based filtering

Creating Profile

```python
Anubhav_movie_ratings = [
            {'title':'Predator', 'rating':4.9},
            {'title':'Final Destination', 'rating':4.9},
            {'title':'Mission Impossible', 'rating':4},
            {'title':"Beverly Hills Cop", 'rating':3},
            {'title':'Exorcist, The', 'rating':4.8},
            {'title':'Waiting to Exhale', 'rating':3.9},
            {'title':'Avengers, The', 'rating':4.5},
            {'title':'Omen, The', 'rating':5.0}
         ]
Anubhav_movie_ratings = pd.DataFrame(Anubhav_movie_ratings)
Anubhav_movie_ratings
```

Add movieId to input user

```python
Anubhav_movie_Id = movies_df[movies_df['title'].isin(Anubhav_movie_ratings['title'])]
Anubhav_movie_ratings = pd.merge(Anubhav_movie_Id, Anubhav_movie_ratings)
Anubhav_movie_ratings
```

Dropping year and genres

```python
Anubhav_movie_ratings = Anubhav_movie_ratings.drop(['genres','year'], 1)
Anubhav_movie_ratings
```

subset of movies that the input has watched from the dataframe containing genres defined with binary values.

```python
Anubhav_genres_df = movies_with_genres[movies_with_genres.movieId.isin(Anubhav_movie_ratings.movieId)]
Anubhav_genres_df
```

dropping the movieId, title, genres and year columns.

```python
Anubhav_genres_df.reset_index(drop=True, inplace=True)
Anubhav_genres_df.drop(['movieId','title','genres','year'], axis=1, inplace=True)
Anubhav_genres_df
```

multiplying movie ratings by Anubhav_genres_df table

```python
Anubhav_profile = Anubhav_genres_df.T.dot(Anubhav_movie_ratings.rating)
Anubhav_profile
```

Content-Based Recommender System

```python
movies_with_genres = movies_with_genres.set_index(movies_with_genres.movieId)
movies_with_genres.head()
```

delete irrelevant columns from the movies_with_genres data frame

```python
movies_with_genres.drop(['movieId','title','genres','year'], axis=1, inplace=True)
movies_with_genres.head()
```

 weighted average of every movie based on his profile and recommend the top twenty movies that match his preference.

```python
recommendation_table_df = (movies_with_genres.dot(Anubhav_profile)) / Anubhav_profile.sum()
recommendation_table_df.head()
```

recommendation table

```python
recommendation_table_df.sort_values(ascending=False, inplace=True)
copy = movies_df.copy(deep=True)
copy = copy.set_index('movieId', drop=True)
top_index = recommendation_table_df.index[:20].tolist()

recommended_movies = copy.loc[top_index, :]
recommended_movies
```
