```python
import pandas as pd
import numpy as np

df= pd.read_csv("./tested.csv")
df.head()
```

One-hot encoding

```python
one_hot = pd.get_dummies(df,columns=['Sex','Embarked'])
one_hot
```

Perform ordinal - encoding: replace the labels by some ordinal number if ordinal is meaningful


```python
ordinal_mapping = {1: 3, 2: 2, 3: 1}

df['Pclass_Ordinal'] = df['Pclass'].map(ordinal_mapping)
print(df.head())
```

Perform Mean encoding: replace the label by the mean of the target for that label


```python
mean_encoded = df.groupby('Embarked')['Survived'].mean()
df['Embarked_MeanEncoded'] = df['Embarked'].map(mean_encoded)
print(df.head())
```

Perform target-encoding: similar to mean encoding, but use both posterior probability and probability of the target

```python
posterior_prob = df.groupby('Embarked')['Survived'].mean()
prob = df['Survived'].mean()
target_encoded = (posterior_prob * len(df) + prob * 10) / (len(df) + 10)
df['Embarked_TargetEncoded'] = df['Embarked'].map(target_encoded)
print(df.head())
```

Perform Weight of Evidence (WOE) encoding: replace the label with WOE of each label.

```python
proportion_survived = df.groupby('Embarked')['Survived'].mean()
proportion_not_survived = 1 - proportion_survived

woe = np.log(proportion_survived / proportion_not_survived)
df['Embarked_WOE'] = df['Embarked'].map(woe)
print(df.head())
```
